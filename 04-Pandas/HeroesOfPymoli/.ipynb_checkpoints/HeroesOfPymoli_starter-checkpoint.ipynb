```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = pd.DataFrame([{"Total Players: " : purchase_data["SN"].nunique()}])
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueitems = purchase_data["Item Name"].nunique()
unique_items

average_price = np.mean(purchase_data["Price"])
average_price

no_of_purchases = purchase_data["Item Name"].size
no_of_purchases

sales = sum(purchase_data["Price"])
sales

Summary = pd.DataFrame([{"No. of Unique Items": unique_items, 
                         "Avg. Price of Purchase": average_price, 
                         "Total No. of Purchases": no_of_purchases, 
                         "Total Sales": sales}])

Summary["Avg. Price of Purchase"] = Summary["Avg. Price of Purchase"].map("${:.2f}".format)
Summary["Total Sales"] = Summary["Total Sales"].map("${:.2f}".format)
Summary = Summary[["No. of Unique Items", "Avg. Price of Purchase", "Total No. of Purchases", "Total Sales"]]
Summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
demo = purchase_data.groupby(['Gender'])
total_count = demo ["SN"].nunique()
total_count

percentage = (total_count/purchase_data["SN"].nunique()) * 100
percentage

summary_pc = pd.DataFrame({"Total Count": total_count, "Player Percentage": percentage})
summary_pc ["Player Percentage"] = summary_pc["Player Percentage"].map("{:.2f}".format)
summary_pc = summary_pc[["Total Count", "Player Percentage"]]
summary_pc
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
demo = purchase_data.groupby(['Gender'])
total_count = demo ["SN"].nunique()
total_count

average_price = np.mean(purchase_data["Price"])
average_price

gender_purchases = demo["Purchase ID"].count()
gender_purchases

gender_total = demo["Price"].sum()
gender_total

gender_avg = gender_total/total_count
gender_avg

summary_gd = pd.DataFrame({"Purchase Count": gender_purchases, "Average Price": average_price,
                          "Total Purchases": gender_total, "Avg. Purchase Per Person": gender_avg})

summary_gd["Average Price"] = summary_gd["Average Price"].map("${:.2f}".format)
summary_gd["Total Purchases"] = summary_gd["Total Purchases"].map("${:.2f}".format)
summary_gd["Avg. Purchase Per Person"] = summary_gd["Avg. Purchase Per Person"].map("${:.2f}".format)
summary_gd = summary_gd[["Purchase Count", "Average Price", "Total Purchases", "Avg. Purchase Per Person"]]

summary_gd
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

demo = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Demographics"] = pd.cut(purchase_data["Age"], bins, labels = demo)
age_demo = purchase_data.groupby("Age Demographics")

total_count = age_demo ["SN"].nunique()
total_count

total_players = purchase_data["SN"].nunique()
total_players


player_percent = (total_count/total_players) * 100
player_percent

summary_ad = pd.DataFrame({"Total Count": total_count, "Player Percent": player_percent})
summary_ad["Player Percent"] = summary_ad["Player Percent"].map("${:.2f}".format)

summary_ad = summary_ad[["Total Count", "Player Percent"]]
summary_ad
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0 , 9, 14, 19, 24, 29, 34, 39, 100]

demo = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Demographics"] = pd.cut(purchase_data["Age"], bins, labels = demo)
age_demo = purchase_data.groupby("Age Demographics")

purchase_democount = age_demo["Purchase ID"].count()
purchase_democount

avgprice = age_demo["Price"].mean()
purchtotal = purchase_democount * avgprice
avgpp = age_demo["Price"].sum()/age_demo["SN"].nunique()

summary_pa = pd.DataFrame({"Purchase Count": purchase_democount,
                         "Avg. Purchase Price": avgprice,
                         "Total Purchase": purchtotal,
                         "Avg. Per Person": avgpp})
summary_pa["Avg. Purchase Price"] = summary_pa["Avg. Purchase Price"].map("${:.2f}".format)
summary_pa["Total Purchase"] = summary_pa["Total Purchase"].map("${:.2f}".format)
summary_pa["Avg. Per Person"] = summary_pa["Avg. Per Person"].map("${:.2f}".format)


summary_pa.head()



```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spend = purchase_data.groupby("SN")
no_purchase = spend["Purchase ID"].count()
no_purchase
total_purchase = spend["Price"].sum()
total_purchase
avgprice = total_purchase/no_purchase
avgprice

summary_ts = pd.DataFrame({"Purchase Count": no_purchase, "Avg. Purchase Price": avgprice, "Total Purchase Value": total_purchase})

summary_ts = summary_ts.sort_values(by="Total Purchase Value", ascending = False)
summary_ts["Avg. Purchase Price"] = summary_ts["Avg. Purchase Price"].map("${:.2f}".format)
summary_ts["Total Purchase Value"] = summary_ts["Total Purchase Value"].map("${:.2f}".format)

summary_ts = summary_ts[["Purchase Count", "Avg. Purchase Price", "Total Purchase Value"]]
summary_ts.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular = purchase_data.groupby(["Item ID", "Item Name"])
no_purchase = popular["Purchase ID"].count()
itemprice = popular["Price"].mean()
totalvalue = no_purchase * itemprice

summary_mp = pd.DataFrame({"Purchase Count": no_purchase, 
                          "Item Price": itemprice,
                          "Total Purchase Value": totalvalue})

summary_mp["Item Price"] = summary_mp ["Item Price"].map("${:.2f}".format)

summary_mp = summary_mp [["Purchase Count", "Item Price", "Total Purchase Value"]]
summary_mp.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popular = purchase_data.groupby(["Item ID", "Item Name"])
no_purchase = popular["Purchase ID"].count()
itemprice = popular["Price"].mean()
totalvalue = no_purchase * itemprice

summary_mp = pd.DataFrame({"Purchase Count": no_purchase, 
                          "Item Price": itemprice,
                          "Total Purchase Value": totalvalue})

summary_mp = summary_mp.sort_values(by="Purchase Count", ascending = False)
summary_mp["Item Price"] = summary_mp ["Item Price"].map("${:.2f}".format)

summary_mp = summary_mp [["Purchase Count", "Item Price", "Total Purchase Value"]]
summary_mp.head()
```
